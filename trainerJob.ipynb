```python
from google.cloud import aiplatform

aiplatform.init(location='us-central1')

PACKAGE_GCS_URI = 'gs://preprocessed_credit_card_data/trainer-0.1.tar.gz'
CONTAINER_URI = 'us-docker.pkg.dev/vertex-ai/training/scikit-learn-cpu.0-23:latest'
STAGING_BUCKET_URI = 'gs://preprocessed_credit_card_data/job_output'
```

## Create custom training job

```python
job = aiplatform.CustomPythonPackageTrainingJob(
    display_name = 'credit-card-model-job',
    python_package_gcs_uri = PACKAGE_GCS_URI,
    python_module_name = 'trainer.training-script',
    container_uri = CONTAINER_URI,
    staging_bucket = STAGING_BUCKET_URI
)
```

```python
job.run(replica_count = 1, machine_type = 'n1-standard-4')
```
